```python
#Unzipping the dataset file
import zipfile
zip_ref = zipfile.ZipFile('/content/wine-quality-dataset.zip', 'r')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df=pd.read_csv('/content/WineQT.csv')
```

```python
df
```

```python
#Dropping the Id column
df = df.drop(columns=['Id'])
```

```python
df
```

```python
X = df.drop('quality', axis=1)
y = df['quality']
```

```python
#splitting data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=99)
```

```python
#Decision Tree classifier
clf = DecisionTreeClassifier(random_state=25)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
# evaluating the model's performance
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
report = classification_report(y_test, y_pred)
print('Classification Report:\n', report)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
max_depths = np.arange(1, 21)
```

```python
# Initializing lists to store accuracy values
train_accuracies = []
test_accuracies = []
```

```python
for depth in max_depths:
    train_pred = clf.predict(X_train)
    train_accuracy = accuracy_score(y_train, train_pred)
    train_accuracies.append(train_accuracy)

    test_pred = clf.predict(X_test)
    test_accuracy = accuracy_score(y_test, test_pred)
    test_accuracies.append(test_accuracy)
```

```python
#accuracy vs. max_depth graph
plt.figure(figsize=(10, 6))
plt.plot(max_depths, train_accuracies, label='Train Accuracy')
plt.plot(max_depths, test_accuracies, label='Test Accuracy')
plt.xlabel('Max Depth of Decision Tree')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Max Depth for Decision Tree')
plt.legend()
plt.grid(True)
plt.show()
```
